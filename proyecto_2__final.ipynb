##PREGUNTA 1

```python
import pandas as pd
import numpy as np
```

```python
ruta = "/content/proyecto_2/assets/Energy Indicators.xls"
energy = pd.read_excel(ruta, skiprows = 17, skipfooter = 38)
# Elimina primera dos columnas
energy = energy.iloc[:,2:]
```

```python
# Renombrar las columnas
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x*1000000 if x != '...'else np.nan)
energy['Energy Supply per Capita'] = energy['Energy Supply per Capita'].apply(lambda x: x if x != "..."else np.nan)
energy.head()
```

```python
# Eliminar parentesis
energy['Country'] = energy['Country'].str.replace(r"\s*\(.*\)", "", regex = True)
energy['Country'] = energy['Country'].str.replace(r"\d+$", "", regex = True)
energy["Country"] = energy["Country"].str.strip()
energy.head()
```

```python
energy["Country"] = energy["Country"].replace({
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region3": "Hong Kong",
    "China2":"China",
    "Japan10":"Japan",
    "France6": "France",
    "Italy9": "Italy",
    "Spain16": "Spain",
    "Autralia1": "Australia"
})
```

```python
ruta_2 = "/content/proyecto_2/assets/world_bank.csv"

# Cargar datos del PBI a un DataFrame
gdp = pd.read_csv(ruta_2, skiprows=4)
```

```python
gdp["Country Name"].replace({
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
},inplace = True)
```

```python
ruta_3 = "/content/proyecto_2/assets/scimagojr-3.xlsx"
scimago = pd.read_excel(ruta_3)
scimago = scimago[scimago["Rank"] <= 15]
```

```python
# Ccombinar archivos
def carga_datos():
  combinado = pd.merge(scimago, energy, how = "inner", left_on = "Country", right_on = "Country")
  combinado = pd.merge(combinado, gdp, how = "inner", left_on = "Country", right_on = "Country Name")
  columnas_requeridas = ["Country", "Rank", "Documents", "Citable documents", "Citations", "Self-citations", "Citations per document", "H index", "Energy Supply", "Energy Supply per Capita", "% Renewable", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015"]
  combinado = combinado[columnas_requeridas]
  combinado.set_index('Country', inplace = True)
  return combinado
carga_datos()
```

##PREGUNTA 2
La pregunta anterior unió tres datasets y luego los redujo a las 15 entradas principales. Pero antes de reducirlos a los 15 elementos principales, ¿cuántas entradas se perdieron?

```python
def outer():
  outer1 = pd.merge(scimago, energy, how = "outer", left_on = "Country", right_on = "Country")
  outer1 = pd.merge(outer1, gdp, how = "outer", left_on = "Country", right_on = "Country Name")
  cantidad = len(outer1)
  cantidad_combinado = len(carga_datos())
  resultado = cantidad - cantidad_combinado
  return resultado
outer()
```

##PREGUNTA 3
¿Cuáles son los 15 países con mayor GDP (PBI) promedio en los últimos 10 años?

```python
def promedio_pbi():
  gdp_columnas = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
  gdp_datos = gdp[gdp_columnas]
  avg = gdp_datos.mean(axis = 1)
  avg1 = gdp[["Country Name"]].copy()
  avg1 ["Average GDP"] = avg
  avg1 = avg1.sort_values(by = "Average GDP",ascending = False)
  top15 = avg1.head(15)
  avgGDP = top15.set_index("Country Name")["Average GDP"]
  return avgGDP
promedio_pbi()
```

##PREGUNTA 4
¿Cuánto cambió el GDP (PBI) durante el lapso de 10 años para el país con el sexto GDP (PBI) promedio más grande?

```python
def GDP_cambios():
  country_6 = promedio_pbi().index[5]
  GDP2006 = gdp[gdp["Country Name"] == country_6]["2006"].values[0]
  GDP2015 = gdp[gdp["Country Name"] == country_6]["2015"].values[0]
  GDP_cambio= GDP2015 - GDP2006
  return GDP_cambio
GDP_cambios()
```

```python
def energy_pC():
  avg= carga_datos()["Energy Supply per Capita"].mean()
  return avg
energy_pC()
```

##PREGUNTA 6
¿Qué país tiene el máximo % Renewable y cuál es el porcentaje?

```python
def porcentaje():
  max_renewable_country = carga_datos()['% Renewable'].idxmax()
  max_renewable_percentage = carga_datos().loc[max_renewable_country, '% Renewable']
  return (max_renewable_country, max_renewable_percentage)
porcentaje()
```

##PREGUNTA 7
Cree una nueva columna que sea la relación entre Self-Citations y Total Citations. ¿Cuál es el valor máximo para esta nueva columna y qué país tiene la relación más alta?

```python
def relacion_columna():
  datos = carga_datos()
  datos["Citations Ratio"] = datos["Self-citations"]/datos["Citations"]
  maxid = datos["Citations Ratio"].idxmax()
  max = datos.loc[maxid, "Citations Ratio"]
  return (maxid, max)
relacion_columna()
```

##PREGUNTA 8
Crea una columna que estime la población utilizando Energy Supply y Energy Supply per capita. ¿Cuál es el tercer país más poblado según esta estimación?

```python
def poblacion():
  poblacion1 = carga_datos()
  poblacion1["Population"] = poblacion1["Energy Supply"] / poblacion1["Energy Supply per Capita"]
  poblacion1 = poblacion1.sort_values(by="Population", ascending = False)
  tercer_pais = poblacion1.iloc[2].name
  return tercer_pais
poblacion()
```

##PREGUNTA 9
Crea una columna que estime el número de Citable documents por persona. ¿Cuál es la correlación entre el número de Citable documents per cápita y el Energy Supply per Capita? Utiliza el método .corr(), (correlación de Pearson).

```python
def citable_document():
  document = carga_datos()
  document["Citable document per Capita"] = document["Citable documents"] / document["Energy Supply per Capita"]
  correlacion= document["Citable documents"].corr(document["Energy Supply per Capita"])
  return correlacion
citable_document()
```

##PREGUNTA 10
Crea una nueva columna con un valor igual a 1 si el valor de % de energía renovable del país (% Renewable) es igual o superior de la mediana de todos los países en el top 15, o con un valor igual a 0 si el valor de % de energía renovable del país (% Renewable) está por debajo de la mediana.

```python
def high_renewable():
    datos = carga_datos()
    top15 = datos.head(15)
    median_renewable = top15['% Renewable'].median()
    datos['HighRenew'] = (datos['% Renewable'] >= median_renewable).astype(int)
    high_renew_series = datos.sort_values(by=['HighRenew', 'Rank'], ascending=[False, True])['HighRenew']
    return high_renew_series
high_renewable()
```

##PREGUNTA 11
Utiliza el siguiente diccionario para agrupar los países por continente, luego crea un DataFrame que muestre el tamaño de la muestra (la cantidad de países en cada contenedor de continente) y la suma, la media y la desviación estándar para la población estimada de cada país.

```python
import pandas as pd

# Crear el diccionario de continentes
continent_dict = {'China':'Asia',
                   'United States':'North America',
                   'Japan':'Asia',
                   'United Kingdom':'Europe',
                   'Russian Federation':'Europe',
                   'Canada':'North America',
                   'Germany':'Europe',
                   'India':'Asia',
                   'France':'Europe',
                   'South Korea':'Asia',
                   'Italy':'Europe',
                   'Spain':'Europe',
                   'Iran':'Asia',
                   'Australia':'Australia',
                   'Brazil':'South America'}

df = pd.DataFrame.from_dict(continent_dict, orient='index', columns=['Continent'])
df['Population'] = carga_datos()['Energy Supply'] / carga_datos()['Energy Supply per Capita']
df_grouped = df.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
df_grouped.columns = ['size', 'sum', 'mean', 'std']
df_grouped
```
